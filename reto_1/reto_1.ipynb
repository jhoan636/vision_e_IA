```python
import time
import numpy as np

ROWS, COLS = 1000, 1000
VECTOR_PATH = "/content/drive/MyDrive/VISION_E_IA/recto_1/vector_1000x1000.txt"

start = time.time()

# 1) leer texto -> array
with open(VECTOR_PATH, "r") as f:
    data_str = f.read().strip()

# convierte a numpy (separado por espacios)
arr = np.fromstring(data_str, sep=' ', dtype=np.uint8)  # uint8 0..255

# 2) estad√≠sticas
min_val = arr.min()
max_val = arr.max()
mean_val = arr.mean()
std_val = arr.std()

end = time.time()

print("Minimo:", int(min_val))
print("Maximo:", int(max_val))
print("Media:", float(mean_val))
print("Desviacion estandar:", float(std_val))
print("Tiempo total (Drive+Numpy):", end - start, "segundos")

# 3) reconstruir imagen 1000x1000
img = arr.reshape(ROWS, COLS)
```

```python
import matplotlib.pyplot as plt
img = arr.reshape(ROWS, COLS) # Cambia a vector
plt.figure(figsize=(6,6)) # Crea la hoja de dibujo
plt.imshow(img, cmap="gray", vmin=0, vmax=255) #Cada numero en un pixel
plt.title("Matriz 1000x1000 como imagen")
plt.axis("off") # Desactivar cuadricula
plt.show() # Visualizar
```
